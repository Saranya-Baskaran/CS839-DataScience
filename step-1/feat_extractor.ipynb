```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stop_words
```

```python
food_adj = pd.read_csv('adjectives.csv')['adjective'].tolist()
food_adj
```

```python
data = pd.read_csv('dataset.csv')
#data.drop(data.columns[-1], axis=1, inplace=True)
data.head(10)
```

```python
def extract_features(data):
    p_doc_id, p_term, adj, wlen, wcount, sword = None, None, [], [], [], []
    for index, row in data.iterrows():
        if p_doc_id and p_doc_id == row[0] and p_term in food_adj:
            adj.append(True)
        else:
            adj.append(False)

        p_doc_id = row[0]
        p_term = row['term']  
        
        wlen.append(len(p_term))
        wcount.append(len(p_term.split(' ')))
        sword.append(p_term in stop_words)

    return adj, wlen, wcount, sword

data['has_leading_food_adj'], data['word_length'], data['word_count'], data['is_stop_word'] = extract_features(data)
```
