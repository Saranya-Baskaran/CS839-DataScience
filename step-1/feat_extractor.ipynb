```python
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
kathy_temean = "https://kathytemean.wordpress.com/2009/04/25/101-descriptive-words-for-foods"
kt_data  = requests.get(kathy_temean).text

fox_hugh = "https://foxhugh.com/word-lists/list-of-food-adjectives/"
fh_data  = requests.get(fox_hugh).text

kt_soup = BeautifulSoup(kt_data)
fh_soup = BeautifulSoup(fh_data, 'html.parser')
```

```python
kt_list = kt_soup.find(id='post-664').find('div', class_='entry').find_all('p')[3].text.split('\n')
kt_list
```

```python
fh_list = [p.text for p in fh_soup.find(id='post-2228').find('div', class_='entry-content').find_all('p')[1:-16]]
fh_list
```

```python
food_adj = set(kt_list + fh_list)
food_adj
```

```python
import pandas as pd
df = pd.DataFrame(list(food_adj))
df.to_csv('adjectives.csv', header=False, index=False)
```

```python
import pandas as pd

food_adj = pd.read_csv('adjectives.csv')['adjective'].tolist()
data = pd.read_csv('dataset_mini.csv')
data.head(10)
```

```python
has_leading_food_adj = []
p_doc_id, p_term = None, None

for index, row in data.iterrows():
    if p_doc_id and p_doc_id == row[0] and p_term in food_adj:
        has_leading_food_adj.append(True)
        print(row['term'] + " preceded by " + p_term)
    else:
        has_leading_food_adj.append(False)
        
    p_doc_id = row[0]
    p_term = row['term']
```

```python
data['has_leading_food_adj'] = has_leading_food_adj
```
